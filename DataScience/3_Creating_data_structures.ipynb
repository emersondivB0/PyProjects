```python
import pandas as pd
```

## Series

```python
# generar una serie pandas
data = [1,2,3,4,5]
s = pd.Series(data)
s
```

```python
# generar nuevos índices
index = ['Line' + str(i) for i in range(5)]
index
```

```python
s = pd.Series(data = data, index = index)
s
```

```python
# también se pueden generar índices a través de un diccionario
data = {'Line' + str(i) : i+1 for i in range(5)}
data
```

```python
s = pd.Series(data)
s
```

```python
# puedo hacer operaciones, como sumarle 2 a cada elemento:
s1 = s + 2
s1
```

```python
# se pueden realizar operaciones entre series:
s2 = s1 + s
s2
```

```python
# puebas:
datos = {'A': {'X': 1, 'Y': 3}, 'B': {'X': 2, 'Y': 4}}
df = pd.DataFrame(datos)
df
```

```python
datos = [('A', 'B'), ('C', 'D')]
df = pd.DataFrame(datos, columns = ['L1', 'L2'],  index = ['C1', 'C2'])
df
```

```python
df1 = pd.DataFrame({'A': {'X': 1}, 'B': {'X': 2}})
df2 = pd.DataFrame({'C': {'X': 3}, 'D': {'X': 4}})
pd.concat([df1, df2])
```

```python
datos = [[1, 2, 3], [4, 5, 6]]
index = 'X,Y'.split(',')
columns = list('CBA')[::-1]
df = pd.DataFrame(datos, index, columns)
df
```

## DataFrame


```python
# First, generate lists:
data = [[1,2,3],[4,5,6],[7,8,9]]
data
```

```python
df1 = pd.DataFrame(data = data)
df1
```

```python
index = ['Line' + str(i) for i in range(3)]
index
```

```python
df1 = pd.DataFrame(data = data, index = index)
df1
```

```python
columns= ['Column' + str(i) for i in range(3)]
columns
```

```python
df1 = pd.DataFrame(data = data, index = index, columns = columns)
df1
```

### With Dictionaries

```python
data = {'Column0' : {'Line0' : 1, 'Line1' : 4, 'Line2' : 7},
        'Column1' : {'Line0' : 2, 'Line1' : 5, 'Line2' : 8},
        'Column2' : {'Line0' : 3, 'Line1' : 6, 'Line2' : 9}}
data
```

```python
df2 = pd.DataFrame(data)
df2
```

### With List of Tuples

```python
data2 = [(1,2,3),(4,5,6),(7,8,9)]
data2
```

```python
df3 = pd.DataFrame(data2, index = index, columns = columns)
df3
```

```python
df1[df1 > 0] = 'A'
df1
```

```python
df2[df2 > 0] = 'B'
df3[df3 > 0] = 'C'
```
