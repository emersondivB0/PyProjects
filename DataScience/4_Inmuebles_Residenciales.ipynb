```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler.csv', sep = ';')
datos
```

```python
datos.head(10)
```

Debemos filtrar los inmuebles residenciales solamente

```python
list(datos['Tipo'].drop_duplicates())
```

```python
residencial = ['Habitación',
 'Casa',
 'Departamento',
 'Casa en condominio',
 'Casa comercial',
 'Casa de villa']
residencial
```

Con esa lista, filtramos los datos que queremos

```python
seleccion = datos['Tipo'].isin(residencial)
seleccion
```

```python
datos_residencial = datos[seleccion]
datos_residencial
```

```python
list(datos_residencial['Tipo'].drop_duplicates())
```

Los índices quedan según el dataframe original, hay que ordenarlos en secuencia

```python
datos_residencial.shape[0]
```
