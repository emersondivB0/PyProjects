```python
import pandas as pd
```

```python
data = [[1,2,3],[4,5,6],[7,8,9]]
```

```python
list('321')
```

```python
df = pd.DataFrame(data, list('321'), list('ZYX'))
df
```

```python
# se puede ordenar con el método sort
df.sort_index(inplace= True)
df
```

```python
# también se ordenan las columnas
df.sort_index(inplace= True, axis= 1)
df
```

```python
# también se pueden ordenar por el contenido
df.sort_values(by = 'X', inplace= True)
df
```

```python
df.sort_values(by = '3', axis= 1, inplace= True)
df
```
