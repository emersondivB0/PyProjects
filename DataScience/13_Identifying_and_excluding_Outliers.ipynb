```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14,6))
```

```python
datos = pd.read_csv('alquiler_residencial.csv', sep=';')
datos.head(10)
```

```python
datos = datos.fillna(0)
datos.head(10)
```

```python
# the boxplot divide the data in quartiles
datos.boxplot(['Valor'])
```

```python
datos[datos['Valor'] >= 500000]
```

```python
valor = datos['Valor']
```

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limit_inf = Q1 - 1.5 * IIQ
limit_sup = Q3 + 1.5 * IIQ
```

```python
selection = ((valor >= limit_inf) & (valor <= limit_sup))
```

```python
datos_new = datos[selection]
```

```python
datos_new.boxplot(['Valor'])
```

```python
datos.hist(['Valor'])
```

```python
datos_new.hist(['Valor'])
```

Now we can group by type of build

```python
datos.boxplot(['Valor'], by = ['Tipo'])
```

```python
datos_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = datos.groupby('Tipo')['Valor']
```

```python
grupo_tipo.groups
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limit_inf = Q1 - 1.5 * IIQ
limit_sup = Q3 + 1.5 * IIQ
```

```python
datos_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo = datos['Tipo'] == tipo
  eh_dentro_limite = (datos['Valor'] >= limit_inf[tipo]) & (datos['Valor'] <= limit_sup[tipo])
  seleccion = eh_tipo & eh_dentro_limite
  datos_seleccion = datos[seleccion]
  datos_new = pd.concat([datos_new, datos_seleccion])
```

```python
datos_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
datos_new.to_csv('alquiler_residencial_sin_outliers.csv', sep=';', index = False)
```
