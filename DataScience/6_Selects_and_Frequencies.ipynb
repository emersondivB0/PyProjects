```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler.csv', sep= ';')
datos
```

## Select only 'Departamento' type

```python
selection = datos['Tipo'] == 'Departamento'
selection
```

```python
n1 = datos[selection].shape[0]
n1
```

## Select 3 types: Casa, Casa en condominio and Casa de Villa

```python
selection1 = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') | (datos['Tipo'] == 'Casa de villa')
selection1
```

```python
n2 = datos[selection1].shape[0]
n2
```

## Select all builds with area between 60 and 100 m2

```python
selection2 = (datos['Area'] >= 60) & (datos['Area'] <= 100)
n3 = datos[selection2].shape[0]
n3
```

## Select builds with 4 rooms and rent less than $2.000,00

```python
selection3 = (datos['Cuartos'] >= 4) & (datos['Valor'] < 2000)
n4 = datos[selection3].shape[0]
n4
```
