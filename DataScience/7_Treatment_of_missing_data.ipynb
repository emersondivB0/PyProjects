# Treatment of missing data

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler.csv', sep=';')
datos.head(10)
```

```python
# first methos to check null data
datos.isnull()
```

```python
datos.notnull()
```

```python
datos.info()
```

```python
# we can see, the column 'Valor' has null values
datos[datos['Valor'].isnull()]
```

```python
# we have to delete those null values rows
A = datos.shape[0]
datos.dropna(subset = ['Valor'], inplace= True)
# with inplace= True, datos change to the new values
B = datos.shape[0]
A-B
```

```python
datos[datos['Valor'].isnull()]
```

Sometimes we have to delete some of the nulls, not all

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
selection = (datos['Mantenimiento'].isnull()) & (datos['Tipo'] == 'Departamento')
```

```python
C = datos.shape[0]
# using ~ you can negate the condition
datos = datos[~selection]
D = datos.shape[0]
C-D
```

```python
# fillna can put a value into NaN
datos = datos.fillna({'Mantenimiento':0, 'Impuesto':0})
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
datos.info()
```

Now we can export

```python
datos.to_csv('alquiler_residencial.cvs', sep=';', index= False)
```
