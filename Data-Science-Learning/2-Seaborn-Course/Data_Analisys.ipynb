## Hypothesis test

```python
sns.lmplot(x='iu', y='valor_factura', data= tarjetas, col='moroso')
```

```python
from scipy.stats import ranksums
```

H<sup>null<sup>

The distribution of moroso and no moroso is the same

H<sup>alt<sup>

The distribution of moroso and no moroso is not the same

```python
moroso = tarjetas.query("moroso == 1").valor_factura
```

```python
moroso
```

```python
no_moroso = tarjetas.query("moroso == 0").valor_factura
```

```python
no_moroso
```

```python
resultado = ranksums(moroso, no_moroso)
```

```python
print("The p_value of our test is {}".format(resultado.pvalue))
```

With this we can accept our alternative hypothesis, about moroso and no moroso aren't the same disttribution
